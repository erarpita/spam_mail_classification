pandas is used to create dataframe(structure our data in frame) as given data set is in csv format that is not easy to analyse, and dataframe is easy to analyse and work with it.

linear_model is a class of the sklearn module if contain different functions for performing machine learning with linear models.
The accuracy_score function computes the accuracy, either the fraction (default) or the count (normalize=False) of correct predictions.




```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and PreProcessing

```python
#loading data from csv to panda dataframe
raw_mail_data=pd.read_csv('/content/mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
#replace the null value with a null string
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

Label Encoding(spam mail as 0 and ham mail as 1),
loc(location)

```python
mail_data.loc[mail_data['Category']=='spam','Category',]=0
mail_data.loc[mail_data['Category']=='ham','Category',]=1
```

```python
#seperating data as texts and label
X = mail_data['Message']
Y=mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
#random_state=3 mean data split into train and test data in same manner
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature Extraction(converting text data to feature vector(numerical value) that can be used as input to logistic regression)



```python
#firstly load the tfidVectorizer
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase='True')
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)

#converting y_test and y_train values in integer  as 0,1 can be taken as string

Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
print(X_train_features)
```

Logistic Regression Model 
(Training the model) 

```python
model=LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

Evaluating the trained model

```python
#prediction on training data

prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
```

```python
print('Accuracy on training data:',accuracy_on_training_data)
```

```python
#prediction on test data

prediction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
```

```python
print('Accuracy on test data:',accuracy_on_test_data)
```

Building a predictive system that predict either the mail given is spam or ham

```python
input_mail=["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"]
#converting text to feature vector
input_mail_features=feature_extraction.transform(input_mail)

#making prediction

prediction=model.predict(input_mail_features)
print(prediction)

if(prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
